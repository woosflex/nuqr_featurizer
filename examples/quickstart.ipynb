# NuXplore Quickstart

This notebook demonstrates a minimal end-to-end usage flow with a synthetic image and nucleus instance map.

```python
import numpy as np
import nuxplore as nf
```

```python
height, width = 64, 64
image = np.zeros((height, width, 3), dtype=np.uint8)

# Add synthetic color structure
for r in range(height):
    for c in range(width):
        image[r, c, 0] = (r * 3 + c * 2) % 256
        image[r, c, 1] = (r * 5 + c * 7) % 256
        image[r, c, 2] = (r * 11 + c * 13) % 256

# Preferred mask format: instance map (0 = background, 1..N = nuclei)
instance_map = np.zeros((height, width), dtype=np.uint32)
instance_map[10:24, 12:26] = 1
instance_map[34:50, 30:46] = 2
```

```python
features = nf.extract_features(image, instance_map, use_gpu=False)
print(f"nuclei: {len(features)}")
print("feature keys:", sorted(features[0].keys())[:10])
features[0]
```
